```python
import numpy as np
from numpy import nan as NA
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(10,4))
df.iloc[1,0] = NA
df.iloc[2:3,2] = NA
df.iloc[5:,3] = NA
df
```

```python
# リストワイズ削除 nanがある行を削除
df.dropna()
```

```python
# ペアワイズ削除 欠損している列を無視して使いたい列だけ利用する。
df[[0,1]].dropna()
```

```python
#fillnaで埋める
df.fillna(0)
```

```python
# 前の値で埋める
df.fillna(method = 'ffill')
```

```python
df.mean()
```

```python
df.fillna(df.mean())
```
