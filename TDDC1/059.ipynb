```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
np.random.seed(0)
```

```python
x = np.random.randn(30)
x
```

```python
y = np.sin(x) + np.random.randn(30)
y
```

```python
plt.figure(figsize=(20,6))
plt.plot(x,y,'o')
```

```python
plt.scatter(x,y)
```

```python
plt.title('Pattern')
plt.xlabel('X')
plt.ylabel('Y')

plt.grid(True)
plt.plot(x,y,'o')
```
