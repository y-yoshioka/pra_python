```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def my_function(x):
    return x ** 2 + 2*x + 1
```

```python
x = np.arange(-10,10)
x
```

```python
plt.figure(figsize = (20,6))
plt.plot(x, my_function(x))
plt.grid(True)
```
