```python
data1 = {
    'id':['100', '101', '102', '103', '104','106', '108', '110', '111', '113'],
    'city':['Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo', 'Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo'],
    'birth_year':[1990, 1989, 1992, 1997, 1982, 1991, 1988, 1990, 1995, 1981],
    'name':['Hiroshi', 'Akiko', 'Yuki', 'Satoru', 'Steeve', 'Mituru', 'Aoi', 'Tarou', 'Suguru', 'Mitsuo']
}
df1 = DataFrame(data1)
df1
```

```python
# データ2の準備
data2 =  {
    'id':['100', '101', '102', '105', '107'],
    'math':[50, 43, 33, 76, 98],
    'english':[90, 30, 20, 50, 30],
    'sex':['M','F', 'F', 'M', 'M'],
    'index_num':[0,1,2,3,4]
}
df2 = DataFrame(data2)
df2
```

```python
# 内部結合
pd.merge(df1,df2, on = 'id')
```

```python
# 全結合
pd.merge(df1, df2, how = 'outer')
```

```python
# indexによるマージ
pd.merge(df1, df2, left_index = True, right_on = 'index_num')
```

```python
# 左外部結合
pd.merge(df1, df2, how='left')
```

```python
city_map = {
    'Tokyo': 'Kanto',
    'Hokkaido': 'Hokkaido',
    'Osaka': 'Kansai',
    'Kyoto':'Kansai'
}
city_map
```

```python
# 参照データを結合
df1['region'] = df1['city'].map(city_map)
df1
```

```python
# birth_year の上の3つの数字・文字を取り出す
df1['up_two_num'] = df1['birth_year'].map(lambda x : str(x)[0:3])
df1
```

```python
# 分割のりゅうど
birth_year_bins = [1980, 1985, 1990, 1995, 2000]
birth_year_cut_data = pd.cut(df1.birth_year, birth_year_bins)
birth_year_cut_data
```

```python
pd.value_counts(birth_year_cut_data)
```

```python
# 数字で分割数指定可能。ここでは2つに分割
pd.cut(df1.birth_year, 2)
```

```python
pd.value_counts(pd.qcut(df1.birth_year, 2))
```

```python
student_data_math = pd.read_csv('chap3/student-mat.csv', sep=';')
student_data_math.head()
```

```python
student_data_math['d2'] = student_data_math.age *2
student_data_math['d3'] = student_data_math['age'].map(lambda x: x*2)
student_data_math
```

```python
absences_bins=[0,1,5,100]
student_data_math_ab_cut_data = pd.cut(student_data_math.absences, absences_bins, right=False)
pd.value_counts(student_data_math_ab_cut_data)
```
