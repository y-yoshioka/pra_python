```python
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
numpy_data_x = np.arange(1000)
numpy_data_x
```

```python
numpy_random_data_y = np.random.randn(1000).cumsum()
numpy_random_data_y
```

```python
plt.figure(figsize = (20,6))
plt.plot(numpy_data_x, numpy_random_data_y, label = 'LLabel')
plt.legend()

plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
```
